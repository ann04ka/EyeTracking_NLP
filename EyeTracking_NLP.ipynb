**Айтрекинг** (или трекинг взгляда) — это технология, которая позволяет отслеживать движение глаз человека и определять, куда именно он смотрит. Айтрекинг применяется в различных областях, включая психологию, маркетинг, медицину, компьютерные интерфейсы и многое другое.

### Процессы зрения: саккады, фиксации и изменения зрачка  

Чтение и визуальное восприятие информации — это динамичный процесс, состоящий из двух основных элементов:

**фиксаций** (пауз, во время которых глаз фокусируется на определенной точке)  и **саккад** (быстрых скачкообразных движений глаз между фиксациями).

Во время фиксации мозг обрабатывает визуальную информацию, а саккады позволяют перенаправлять взгляд на новые участки изображения.

Интересно, что в момент фиксации диаметр зрачка может меняться:

**сужение** (синий круг) часто связано с концентрацией внимания,

**расширение** (красный круг) — с когнитивной нагрузкой или эмоциональной реакцией,

а **стабильный размер** (зеленый круг) указывает на нейтральное восприятие.

На основе существующих алгоритмов айтрекинга можно сформировать изображения, отображающие траекторию взгляда человека при просмотре той или иной информации. В качестве фона используется слайд с экрана монитора во время обследования, поверх которого отображаются фиксации и саккады.

**Цвет круга** отражает изменение зрачка: синий — сужение, красный — расширение, зеленый — отсутствие изменений.

**Размер круга** соответствует длительности фиксации: чем дольше задержка взгляда, тем больше окружность.

**Выбор ответа** осуществляется длительной фиксацией (большой круг).

 Например, расширение зрачка (красный круг) на слове **"спокойно"** может говорить о повышенном интересе или удивлении, а сужение (синий) на слове **"компьютера"** — о сосредоточенности. Финальная фиксация на вопросе **"вслух?"** с расширенным зрачком (красный) указывает на завершающую эмоциональную или когнитивную реакцию. ### Задача формирования описаний траектории взгляда

Задача заключается в преобразовании визуальных данных айтрекинга (фиксаций, саккад, изменений диаметра зрачка) в структурированное текстовое описание на естественном языке. Ключевые аспекты задачи включают:

1. **Визуально-текстовое преобразование**: Точное соотнесение элементов айтрекинга (круги, стрелки) с конкретными областями визуального стимула (текст, интерфейс).

2. **Семантическая интерпретация**: Анализ паттернов движения глаз в контексте представленной информации (например, реверсивные саккады как показатель сложности восприятия).

3. **Количественно-качественный анализ**: Определение значимых характеристик:
   - Продолжительность фиксаций
   - Направление и амплитуда саккад
   - Динамика изменения зрачка

4. **Генерация выводов**: Формулировка интерпретаций наблюдаемых паттернов для различных прикладных областей (UX-дизайн, медицинская диагностика, психологические исследования).

### Преимущества использования языковых моделей (Image2Text или VLM)

1. **Автоматизация комплексного анализа**:
   - Использование языковых моделей обеспечивает:
     * Одновременную обработку изображения экрана и данных трекинга
     * Семантическую привязку движений глаз к содержанию

2. **Контекстно-зависимая интерпретация**:
   - Автоматическое определение значимости элементов интерфейса
   - Выявление корреляций между характеристиками взгляда и:
     * Эмоциональными реакциями (по динамике зрачка)
     * Когнитивной нагрузкой (по паттернам фиксаций)

3. **Преимущества перед традиционными методами**:
   - **Скорость**: Автоматическая обработка данных
   - **Масштабируемость**: Анализ больших массивов данных
   - **Объективность**: Минимизация субъективной интерпретации
   - **Гибкость**: Адаптация к различным типам визуальных стимулов

4. **Функциональные возможности**:
   - Генерация стандартизированных отчетов
   - Выявление аномальных паттернов (для медицинской диагностики)
   - Сравнительный анализ поведенческих паттернов

```python
import os
import shutil

def extract_archive(input_archive, output_directory):
    """
    Разархивирует ZIP-архив в указанную директорию.

    :param input_archive: Путь к ZIP-архиву.
    :param output_directory: Путь к директории для разархивации.
    """
    if not os.path.exists(input_archive):
        raise FileNotFoundError(f"Архив не найден: {input_archive}")

    try:
        shutil.unpack_archive(input_archive, output_directory)

        print(f"Архив успешно разархивирован в: {output_directory}")
    except Exception as e:
        print(f"Ошибка при разархивации: {e}")


output_archive = "/content/drive/MyDrive/data/scanpaths_dataset_archive.zip"
extraction_directory = "/content/scanpaths_dataset/"

extract_archive(output_archive, extraction_directory)
```

```python
import os
from datasets import Dataset, DatasetDict
from PIL import Image
import numpy as np
import json

def load_custom_dataset(data_dir):
    """
    Загружает данные из указанной директории и формирует кастомный датасет с изображениями и captions.

    :param data_dir: Путь к директории с данными.
    :return: Объект DatasetDict с train, eval и test разделами.
    """
    images = []
    captions = []

    for filename in os.listdir(data_dir):
        file_path = os.path.join(data_dir, filename)

        if filename.endswith(".json"):
            with open(file_path, "r", encoding="utf-8") as f:
                try:
                    data_items = json.load(f)

                    if isinstance(data_items, list):
                        for data in data_items:
                            image_path = data.get("image_path", "")
                            caption = data.get("caption", "")

                            image_path = "/content/scanpaths_dataset/resized/"+ image_path[image_path.find('/', 15):]
                            if not os.path.exists(image_path):
                                print(f"Изображение не найдено для файла {filename}: {image_path}")
                                continue

                            image = Image.open(image_path).convert("RGB")
                            image_array = image

                            images.append(image_array)
                            captions.append(caption)
                    elif isinstance(data_items, dict):
                        image_path = data_items.get("image_path", "")
                        caption = data_items.get("caption", "")

                        image_path = "/content/scanpaths_dataset/resized/"+ image_path[image_path.find('/', 15):]
                        if not os.path.exists(image_path):
                            print(f"Изображение не найдено для файла {filename}: {image_path}")
                            continue

                        image = Image.open(image_path).convert("RGB")
                        image_array = image

                        images.append(image_array)
                        captions.append(caption)
                    else:
                        print(f"Неподдерживаемый формат данных в файле {filename}.")
                except json.JSONDecodeError as e:
                    print(f"Ошибка при чтении файла {filename}: {e}")

    dataset = Dataset.from_dict({
        "image": images,
        "caption": captions
    })

    train_test_split = dataset.train_test_split(test_size=0.1, shuffle=True, seed=42)
    train_val_split = train_test_split["train"].train_test_split(test_size=0.1, shuffle=True, seed=42)

    custom_dataset = DatasetDict({
        "train": train_val_split["train"],
        "eval": train_val_split["test"],
        "test": train_test_split["test"]
    })

    return custom_dataset


data_directory = "/content/scanpaths_dataset/"  # Путь к директории с данными

custom_dataset = load_custom_dataset(data_directory)

print(custom_dataset)
```

```python
from collections import Counter
import numpy as np
from datasets import DatasetDict
import nltk

def compute_metrics(dataset_dict):
    stats = {}
    vocab_counter = Counter()

    for split in dataset_dict.keys():
        dataset = dataset_dict[split]
        descriptions = [example["caption"] for example in dataset]

        image_count = len(descriptions)

        token_lengths = [len(desc.split()) for desc in descriptions]
        avg_length = np.mean(token_lengths)

        for desc in descriptions:
            tokens = desc.lower().split()
            vocab_counter.update(tokens)

        stats[split] = {
            "image_count": image_count,
            "avg_length": avg_length
        }

    vocab_size = len(vocab_counter)

    return stats, vocab_size

stats, vocab_size = compute_metrics(custom_dataset)

print(f"Images: Train {stats['train']['image_count']}, Eval {stats['eval']['image_count']}, Test {stats['test']['image_count']}")
print(f"Avg. Description Length: Train {stats['train']['avg_length']:.1f}, Eval {stats['eval']['avg_length']:.1f}, Test {stats['test']['avg_length']:.1f}")
print(f"Vocabulary Size: {vocab_size}")
```

```python
import re

system_message = """You are an image recognition AI that provides detailed descriptions of images on russian"""

def extract_question(description):
    pattern = r"\*\*Текст вопроса:\*\*\s*(.*?)\n"

    match = re.search(pattern, description.get('caption', ''), re.DOTALL)

    if match:
        return match.group(1).strip()
    else:
        print('ОТСУТСТВУЕТ ВОПРОС')
        return None

def extract_answer(description):
    pattern = r'Человек выбрал ответ "(Да|Нет)"'

    match = re.search(pattern, description.get('caption', ''), re.DOTALL)

    if match:
        return match.group(1).strip()  # Возвращает "Да" или "Нет"
    else:
        print('ОТСУТСТВУЕТ ОТВЕТ')
        return None

def get_query(description):
    question = extract_question(description)
    answer = extract_answer(description)

    return f'''Опиши траекторию перемещения взгляда по следующему шаблону, если:

Текст вопроса: "{question}"
Выбранный ответ: "{answer}"

1. Текст вопроса: {question}

2. Начало движения:
- Где начинается движение взгляда? Например: «Взгляд начинает движение с первого слова вопроса.»

3. Изменение диаметра зрачка:
- На каких словах или фрагментах текста наблюдалось сужение зрачка (синий круг)?
- На каких словах или фрагментах текста наблюдалось расширение зрачка (красный круг)?
- На каких словах или фрагментах текста диаметр зрачка оставался средним (зеленый круг)?

4. Движение взгляда:
- Опиши последовательность движения взгляда по тексту. Например: «Взгляд движется от начала вопроса вправо по строке, затем переходит ко второй строке.»
- Укажи реверсивные саккады (красные стрелки), если они есть.
- Укажи прямые саккады (синие стрелки), если они есть.
- Укажи переключение между ответами (зеленые стрелки), если оно наблюдалось.

5. Выбор ответа:
- Какой ответ выбрал человек? Например: «Человек выбрал ответ "{answer}" после длительной фиксации на этом варианте.»

Важное правило:
- Запрещено начинать описание фразами типа: «Конечно, вот описание», «Давайте проанализируем траекторию», «Начнем с того, что» или любыми другими вводными фразами. Описание должно быть кратким и начинаться непосредственно с фактической информации.

Легенда:
- Синий круг: Суженный зрачок. Обозначает, что во время фиксации зрачок сужался.
- Зеленый круг: Средний зрачок. Обозначает, что диаметр зрачка не изменялся во время фиксации.
- Красный круг: Расширенный зрачок. Обозначает, что зрачок расширялся во время фиксации.
- Красная стрелка: Реверсивная саккада. Обозначает движение взгляда, которое возвращается к предыдущей точке фиксации.
- Синяя стрелка: Прямая саккада. Обозначает прямое движение взгляда от одной точки фиксации к другой.
- Зеленая стрелка: Переключение между ответами. Обозначает переключение взгляда между различными вариантами ответов.
- Маленький круг: Короткая фиксация. Обозначает короткую по продолжительности фиксацию взгляда.
- Большой круг: Длительная фиксация. Обозначает длительную по продолжительности фиксацию взгляда.
- Выбор ответа осуществляется длительной фиксацией взгляда в области соответствующего варианта.'''
```

```python
def remove_image_path(text):
    """
    Удаляет путь до картинки из текста.

    Параметры:
        text (str): Исходный текст, содержащий путь до картинки в начале.

    Возвращает:
        str: Текст без пути до картинки.
    """
    newline_index = text.find('\n')

    if newline_index != -1:
        return text[newline_index + 1:].strip()

    return text.strip()

text = '../data_anya/resized/5_105.png\n1. **Текст вопроса:**\n - Я какой-либо материальный ущерб компании за время работы наносил?\n\n2. **Начало движения:**\n - Взгляд начинает движение с первого слова вопроса: "Я".'
cleaned_text = remove_image_path(text)
print(cleaned_text)
```

# Metric

```python
import re
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

class AnswerSimilarityAnalyzer:
    def __init__(self):
        self.sbert_model = SentenceTransformer('all-MiniLM-L6-v2')

        self.answer_keywords = {
            'да': 'yes',
            'нет': 'no',
            'Да': 'yes',
            'Нет': 'no'
        }


    def extract_answer(self, text):
        """Извлекает последний встречающийся ответ из текста, проверяя с конца к началу."""
        text = text.replace('"', '')
        reversed_text = ' '.join(text.lower().split()[::-1])[:100]

        for pattern, answer in self.answer_keywords.items():
            if re.search(rf'\b{pattern}\b', reversed_text):
                # print(answer, reversed_text)
                return answer

    def compare_texts(self, text1, text2):
        """Сравнивает два текста по семантике и ответам"""
        answer1 = self.extract_answer(text1)
        answer2 = self.extract_answer(text2)

        embeddings = self.sbert_model.encode([text1, text2])
        semantic_sim = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]

        if answer1 and answer2:
            answer_sim = 1.0 if answer1 == answer2 else 0.0
            answers_match = answer1 == answer2
        else:
            answer_sim = 0.0
            answers_match = None

        total_sim = 0.6 * semantic_sim + 0.4 * answer_sim

        return {

            'semantic_similarity': float(semantic_sim),
            'answer_similarity': float(answer_sim),
            'total_similarity': total_sim,
            'answers_match': answers_match,
            'answers': (answer1, answer2)
        }
```

```python
import json

def write_list_of_lists_to_json(filename, list_of_lists):
    """
    Записывает список списков в JSON-файл.

    :param filename: Имя файла, в который нужно записать JSON.
    :param list_of_lists: Список списков для записи.
    """
    with open(filename, 'w', encoding='utf-8') as file:
        json.dump(list_of_lists, file, ensure_ascii=False, indent=4)

def read_list_of_lists_from_json(filename):
    """
    Читает список списков из JSON-файла.

    :param filename: Имя файла, из которого нужно прочитать JSON.
    :return: Список списков.
    """
    with open(filename, 'r', encoding='utf-8') as file:
        return json.load(file)
```

```python
import torch
import requests
from PIL import Image
from transformers import AutoProcessor, AutoModelForImageTextToText

version = 'BLIP'
processor = AutoProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = AutoModelForImageTextToText.from_pretrained("Salesforce/blip-image-captioning-base", torch_dtype=torch.float16).to("cuda")

results = []

for i in range(len(test_dataset)):
    raw_image = test_dataset[0][1]['content'][0]['image'].convert('RGB')
    inputs = processor(raw_image, return_tensors="pt").to("cuda", torch.float16)
    out = model.generate(**inputs)

    output_text = [processor.decode(out[0], skip_special_tokens=True)]
    results.append(output_text)
```

```python
def collate_fn(examples):
    # Get the texts and images, and apply the chat template
    texts = [
        processor.apply_chat_template(example, tokenize=False) for example in examples
    ]  # Prepare texts for processing
    image_inputs = [process_vision_info(example)[0] for example in examples]  # Process the images to extract inputs

    # Tokenize the texts and process the images
    batch = processor(
        text=texts, images=image_inputs, return_tensors="pt", padding=True
    )  # Encode texts and images into tensors

    # The labels are the input_ids, and we mask the padding tokens in the loss computation
    labels = batch["input_ids"].clone()  # Clone input IDs for labels
    labels[labels == processor.tokenizer.pad_token_id] = -100  # Mask padding tokens in labels

    # Ignore the image token index in the loss computation (model specific)
    if isinstance(processor, AutoProcessor):  # Check if the processor is Qwen2VLProcessor
        image_tokens = [151652, 151653, 151655]  # Specific image token IDs for Qwen2VLProcessor
    else:
        image_tokens = [processor.tokenizer.convert_tokens_to_ids(processor.image_token)]  # Convert image token to ID

    # Mask image token IDs in the labels
    for image_token_id in image_tokens:
        labels[labels == image_token_id] = -100  # Mask image token IDs in labels

    batch["labels"] = labels  # Add labels to the batch

    return batch  # Return the prepared batch
```

```python
from peft import LoraConfig, get_peft_model

# Configure LoRA
peft_config = LoraConfig(
    lora_alpha=6,
    lora_dropout=0.1,
    r=6,
    bias="none",
    target_modules=[
        # Визуальная часть: механизмы внимания в последних 4 слоях SiglipEncoder
        "vision_tower.vision_model.encoder.layers.23.self_attn.q_proj",
        "vision_tower.vision_model.encoder.layers.23.self_attn.k_proj",
        "vision_tower.vision_model.encoder.layers.23.self_attn.v_proj",
        "vision_tower.vision_model.encoder.layers.23.self_attn.out_proj",
        "vision_tower.vision_model.encoder.layers.24.self_attn.q_proj",
        "vision_tower.vision_model.encoder.layers.24.self_attn.k_proj",
        "vision_tower.vision_model.encoder.layers.24.self_attn.v_proj",
        "vision_tower.vision_model.encoder.layers.24.self_attn.out_proj",
        "vision_tower.vision_model.encoder.layers.25.self_attn.q_proj",
        "vision_tower.vision_model.encoder.layers.25.self_attn.k_proj",
        "vision_tower.vision_model.encoder.layers.25.self_attn.v_proj",
        "vision_tower.vision_model.encoder.layers.25.self_attn.out_proj",
        "vision_tower.vision_model.encoder.layers.26.self_attn.q_proj",
        "vision_tower.vision_model.encoder.layers.26.self_attn.k_proj",
        "vision_tower.vision_model.encoder.layers.26.self_attn.v_proj",
        "vision_tower.vision_model.encoder.layers.26.self_attn.out_proj",

        # Текстовая часть: механизмы внимания и MLP в последних 4 слоях Gemma3DecoderLayer
        "language_model.model.layers.30.self_attn.q_proj",
        "language_model.model.layers.30.self_attn.k_proj",
        "language_model.model.layers.30.self_attn.v_proj",
        "language_model.model.layers.30.self_attn.o_proj",
        "language_model.model.layers.31.self_attn.q_proj",
        "language_model.model.layers.31.self_attn.k_proj",
        "language_model.model.layers.31.self_attn.v_proj",
        "language_model.model.layers.31.self_attn.o_proj",
        "language_model.model.layers.32.self_attn.q_proj",
        "language_model.model.layers.32.self_attn.k_proj",
        "language_model.model.layers.32.self_attn.v_proj",
        "language_model.model.layers.32.self_attn.o_proj",
        "language_model.model.layers.33.self_attn.q_proj",
        "language_model.model.layers.33.self_attn.k_proj",
        "language_model.model.layers.33.self_attn.v_proj",
        "language_model.model.layers.33.self_attn.o_proj",

        "language_model.model.layers.30.mlp.gate_proj",
        "language_model.model.layers.30.mlp.up_proj",
        "language_model.model.layers.30.mlp.down_proj",
        "language_model.model.layers.31.mlp.gate_proj",
        "language_model.model.layers.31.mlp.up_proj",
        "language_model.model.layers.31.mlp.down_proj",
        "language_model.model.layers.32.mlp.gate_proj",
        "language_model.model.layers.32.mlp.up_proj",
        "language_model.model.layers.32.mlp.down_proj",
        "language_model.model.layers.33.mlp.gate_proj",
        "language_model.model.layers.33.mlp.up_proj",
        "language_model.model.layers.33.mlp.down_proj",

        # Голова языковой модели
        "language_model.lm_head"
    ],
    task_type="CAUSAL_LM",
)

# Apply PEFT model adaptation
peft_model = get_peft_model(model, peft_config)

# Print trainable parameters
peft_model.print_trainable_parameters()
```

```python
from trl import SFTConfig

training_args = SFTConfig(
    output_dir="Gemma_3_4b_v2/",  # Directory to save the model
    num_train_epochs=2,  # Number of training epochs
    per_device_train_batch_size=1,  # Batch size for training
    per_device_eval_batch_size=1,  # Batch size for evaluation
    gradient_accumulation_steps=4,  # Steps to accumulate gradients
    gradient_checkpointing=True,  # Enable gradient checkpointing for memory efficiency
    # Optimizer and scheduler settings
    optim="adamw_8bit",  # Optimizer type
    learning_rate=2e-4,  # Learning rate for training
    lr_scheduler_type="constant",  # Type of learning rate scheduler
    # Logging and evaluation
    logging_steps=10,  # Steps interval for logging
    eval_steps=10,  # Steps interval for evaluation
    eval_strategy="steps",  # Strategy for evaluation
    save_strategy="steps",  # Strategy for saving the model
    save_steps=10,  # Steps interval for saving
    metric_for_best_model="eval_loss",  # Metric to evaluate the best model
    greater_is_better=False,  # Whether higher metric values are better
    load_best_model_at_end=True,  # Load the best model after training
    # Mixed precision and gradient settings
    bf16=True,  # Use bfloat16 precision
    tf32=False,  # Use TensorFloat-32 precision
    max_grad_norm=0.3,  # Maximum norm for gradient clipping
    warmup_ratio=0.03,  # Ratio of total steps for warmup
    # Hub and reporting
    push_to_hub=True,  # Whether to push model to Hugging Face Hub
    report_to="wandb",  # Reporting tool for tracking metrics
    # Gradient checkpointing settings
    gradient_checkpointing_kwargs={"use_reentrant": False},  # Options for gradient checkpointing
    # Dataset configuration
    dataset_text_field="",  # Text field in dataset
    dataset_kwargs={"skip_prepare_dataset": True},  # Additional dataset options
    # max_seq_length=1024  # Maximum sequence length for input
)

training_args.remove_unused_columns = False  # Keep unused columns in dataset
```

```python
def collate_fn(examples):
    # Get the texts and images, and apply the chat template
    texts = [
        processor.apply_chat_template(example, tokenize=False) for example in examples
    ]  # Prepare texts for processing
    image_inputs = [process_vision_info(example)[0] for example in examples]  # Process the images to extract inputs

    # Tokenize the texts and process the images
    batch = processor(
        text=texts, images=image_inputs, return_tensors="pt", padding=True
    )  # Encode texts and images into tensors

    # The labels are the input_ids, and we mask the padding tokens in the loss computation
    labels = batch["input_ids"].clone()  # Clone input IDs for labels
    labels[labels == processor.tokenizer.pad_token_id] = -100  # Mask padding tokens in labels

    # Ignore the image token index in the loss computation (model specific)
    if isinstance(processor, AutoProcessor):  # Check if the processor is Qwen2VLProcessor
        image_tokens = [151652, 151653, 151655]  # Specific image token IDs for Qwen2VLProcessor
    else:
        image_tokens = [processor.tokenizer.convert_tokens_to_ids(processor.image_token)]  # Convert image token to ID

    # Mask image token IDs in the labels
    for image_token_id in image_tokens:
        labels[labels == image_token_id] = -100  # Mask image token IDs in labels

    batch["labels"] = labels  # Add labels to the batch

    return batch  # Return the prepared batch
```

```python
import shutil
import os

def archive_folder(folder_path, output_filename, archive_format='zip'):
    """
    Архивирует указанную папку.

    :param folder_path: Путь к папке, которую нужно архивировать.
    :param output_filename: Имя выходного архива (без расширения).
    :param archive_format: Формат архива ('zip', 'tar', 'gztar', 'bztar', 'xztar').
    """
    if not os.path.exists(folder_path):
        print(f"Ошибка: Папка {folder_path} не существует.")
        return

    output_path = os.path.join(os.path.dirname(folder_path), output_filename)

    try:
        print(f"Архивация папки {folder_path} в формате {archive_format}...")
        shutil.make_archive(output_path, archive_format, folder_path)
        print(f"Архив успешно создан: {output_path}.{archive_format}")
    except Exception as e:
        print(f"Ошибка при архивации: {e}")

trainer.save_model('/content/data/Gemma_4b_v2')
folder_to_archive = "/content/data/Gemma_4b_v2"
output_archive_name = "/content/drive/MyDrive/data/Gemma_3_4b_v2_archive"
archive_format = "zip"

archive_folder(folder_to_archive, output_archive_name, archive_format)
```
